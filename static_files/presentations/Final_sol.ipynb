3. When evaluating the `Ifraction` object, show the fraction number as a text string like: `The numerator is 1 and the denominator is 2`.

Hint: You may use the Greatest Common Divisor from the `math` or `NumPy` module to simplify the fraction. In addition, be aware of how you access the private attributes from the parent class.

````python
# Do not modify the code below
class Fraction():
    def __init__(self, numerator, denominator):
        self.set_num_den(numerator, denominator)
    def set_num_den(self, numerator, denominator): #setter
        self.__num = numerator
        self.__den = denominator
    def get_num_den(self): #getter
        return self.__num, self.__den
    def __add__(self, otherfrac):
        newnum = self.__num * otherfrac.__den + self.__den* otherfrac.__num
        newden = self.__den * otherfrac.__den
        return Fraction(newnum, newden)
    def __str__(self):
        num, den = self.get_num_den()        
        return f"{num}/{den}"
# Do not modify the code above
    
class Ifraction(Fraction):
    def __init__(self, numerator, denominator):
        super().__init__(numerator, denominator)
    def __mul__(self, otherfrac):
        num, den = super().get_num_den()   
        num2, den2 = otherfrac.get_num_den() 
        newnum = num * num2
        newden = den * den2
        g = np.gcd(newnum, newden)
        return Ifraction(int(newnum/g), int(newden/g))
    def __str__(self):
        num, den = super().get_num_den()        
        return f"{num}/{den}"
    def __repr__(self):
        num, den = super().get_num_den()
        return f"The numerator is {num} and the denominator is {den}"
````
